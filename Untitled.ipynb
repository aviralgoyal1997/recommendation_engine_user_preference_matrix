```python
import pandas as pd
import numpy as np
```

```python
#Reading users file:
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('u.user', sep='|', names=u_cols,
 encoding='latin-1')

#Reading ratings file:
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('u.data', sep='\t', names=r_cols,
 encoding='latin-1')

#Reading items file:
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('u.item', sep='|', names=i_cols,
 encoding='latin-1')
```

Now to get alist of movies rated by  user 

```python
each_user_rating=list(ratings.groupby('user_id')['movie_id']) #list of movies watched by each user
```

```python
each_user_movie_ratings=list(ratings.groupby('user_id')['rating']) #list of ratings given by each user
```

list of movies watched by user 1 

```python
qw=each_user_rating[0][1]
```

```python
aw=each_user_movie_ratings[0][1] #ratings given by user 1 
```

```python
#Getting details of movies watched by user1
a=pd.DataFrame(columns=i_cols)
j=0
for i in qw:
    a.loc[items.index[j]] = items.iloc[i-1]
    j=j+1
```

```python
#Now we need to find preference given by user1 to each genre so
preference_cols=i_cols[6:]
preference={}
for i in i_cols[6:]:
    preference[i]=(a[i].sum()/len(a))
```

Above method to find preference is good but there might be a case when user have watched more movies of genre A than B but acc to his ratings he gave most of A movies 2 star while B movies 4 star so it means he gave more preference to B so we gonna use weighted average for preference 

```python
weighted_preference_cols=i_cols[6:]
weighted_preference={}
for i in i_cols[6:]:
    weighted_preference[i]=(a[i]*aw).sum()/sum(aw) #before multipying make sure index of both series are same
```
